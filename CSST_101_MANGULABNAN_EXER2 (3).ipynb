```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
trials = 1000
head = 0.5
distribution = np.random.binomial(n = 1, p = head, size = trials)

plt.hist(distribution, bins = 2)
plt.title('Binomial Distribution')
plt.show()
```

Follow up:

p = head = 0.5: This is the probability of success. At 0.5, it's equally likely to get a 1 or a 0. As you increase p closer to 1, more 1s will dominate the distribution, and as you lower p towards 0, more 0s will dominate.

trials = 1000: This is the number of times the random sampling is done. With more trials, your histogram will look smoother, and the proportions of 0s and 1s will better reflect the probability distribution.
